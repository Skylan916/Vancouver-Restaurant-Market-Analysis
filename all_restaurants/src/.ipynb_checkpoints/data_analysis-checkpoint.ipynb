```python
import folium
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
import seaborn
seaborn.set()
```

Vancouver Neighbours

```python
m = folium.Map()
regions = pd.read_csv('../data/local-area-boundary.csv')

def plot_region(r):
    folium.Circle(
        location = [r['lat'], r['lng']],
        radius = 100,
        popup = ('{}, Vancouver, BC'.format(r['name'])),
        color = '#f54542',
        fill_color = '#f54542'
    ).add_to(m)

regions.apply(plot_region, axis=1)
folium.LatLngPopup().add_to(m)
m.fit_bounds(m.get_bounds())
display(m)
m.save('../output/regions.html')
```

Vancouver Neighbourhood Mean Income

```python
census = pd.read_csv('../data/census2016.csv', comment='#')
census.sort_values(by=['average_income'], inplace=True)

fig = plt.figure(figsize=(6,4))
plt.bar(census['region'], census['average_income'])
plt.title('Vancouver Neighbourhood Mean Income (2016)')
plt.xlabel('Neighbourhood')
plt.ylabel('Mean Income (CAD)')
plt.xticks(rotation=90)
plt.show()
fig.savefig('../output/mean_income.jpg', dpi=fig.dpi)
```

Vancouver Neighbourhood Population

```python
census.sort_values(by=['total_population'], inplace=True)

fig = plt.figure(figsize=(6,4))
plt.bar(census['region'], census['total_population'])
plt.title('Vancouver Neighbourhood Population (2016)')
plt.xlabel('Neighbourhood')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.show()
fig.savefig('../output/mean_income.jpg', dpi=fig.dpi)
```

Vancouver Neighbourhood Popular Cuisines

```python
restaurants = pd.read_csv('../data/data_clean/restaurants.csv',encoding='unicode_escape')
restaurants.head()
```

```python
restaurants = restaurants.groupby(['region','cuisine']).size().reset_index(name='count')
restaurants.head()
```

```python
preference_by_region = []
regions = census['region'].to_list()

for region in regions:
    cuisine_count = restaurants[restaurants['region'] == region]
    cuisine_count = cuisine_count.sort_values(by=['count'], ascending=False)
    top_cuisines = [region]
    top_cuisines = top_cuisines + cuisine_count['cuisine'].to_list()[:5]
    preference_by_region.append(top_cuisines)

preference_by_region =  pd.DataFrame(preference_by_region, columns=['neighbourhood', 'first', 'second', 'third', 'fourth', 'fifth'])
preference_by_region
```
